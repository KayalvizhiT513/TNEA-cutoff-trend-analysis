```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(10, 6))

# Scatter plots
plt.scatter(cse_df['PUBLIC PERCEPTION RANK'], cse_df['OC 2022'], label='General', alpha=0.7)
plt.scatter(cutoff_7_5_df2022['PP RANK'], cutoff_7_5_df2022['OC'], label='7.5% Reservation', alpha=0.7)

# Regression line for General
x_gen = pd.to_numeric(cse_df['PUBLIC PERCEPTION RANK'], errors='coerce')
y_gen = pd.to_numeric(cse_df['OC 2022'], errors='coerce')
mask = x_gen.notna() & y_gen.notna()
x_gen = x_gen[mask]
y_gen = y_gen[mask]

coef_gen = np.polyfit(x_gen, y_gen, 1)
poly1d_gen = np.poly1d(coef_gen)
plt.plot(x_gen, poly1d_gen(x_gen), color='blue', linestyle='--', label='General Trend')

# Regression line for 7.5% Reservation
# Ensure numeric data
x_res = pd.to_numeric(cutoff_7_5_df2022['PP RANK'], errors='coerce')
y_res = pd.to_numeric(cutoff_7_5_df2022['OC'], errors='coerce')

# Drop rows where either x or y is NaN
mask = x_res.notna() & y_res.notna()
x_res = x_res[mask]
y_res = y_res[mask]

# Now fit regression
coef_res = np.polyfit(x_res, y_res, 1)
poly1d_res = np.poly1d(coef_res)
plt.plot(x_res, poly1d_res(x_res), color='orange', linestyle='--', label='7.5% Reservation Trend')

# Extract coefficients
m1, c1 = coef_gen
m2, c2 = coef_res

# Avoid division by zero
if m1 != m2:
    x_intersect = (c2 - c1) / (m1 - m2)
    y_intersect = m1 * x_intersect + c1

    # Plot the intersection point
    plt.scatter(x_intersect, y_intersect, color='red', zorder=5, label='Intersection Point')
    plt.annotate(f'({x_intersect:.1f}, {y_intersect:.1f})',
                 (x_intersect, y_intersect),
                 textcoords="offset points",
                 xytext=(10, 10),
                 ha='left',
                 color='red',
                 fontsize=9)

# Labels and layout
plt.xlabel('Public Perception Rank')
plt.ylabel('OC Closing Rank')
plt.title('OC Closing Ranks for General and 7.5% Reservation Counseling - 2022')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

plt.figure(figsize=(12, 8))

# Define year-wise data
years = ['2022', '2023', '2024']
general_colors = ['#1f77b4', '#5fa2d0', '#a6cbe5'][::-1]
res_colors = ['#ff7f0e', '#ffb74d', '#ffe0b2'][::-1]
intersect_colors = ['#a626aa', '#d62448', '#0600e0']

for i, year in enumerate(years):
    # General
    x_gen = pd.to_numeric(cse_df['PUBLIC PERCEPTION RANK'], errors='coerce')
    y_gen = pd.to_numeric(cse_df[f'OC {year}'], errors='coerce')
    mask = x_gen.notna() & y_gen.notna()
    xg = x_gen[mask]
    yg = y_gen[mask]
    coef_gen = np.polyfit(xg, yg, 1)
    poly1d_gen = np.poly1d(coef_gen)
    plt.plot(xg, poly1d_gen(xg), linestyle='--', color=general_colors[i],
             label=f'General Trend {year}')

    # Reservation
    res_df = globals()[f'cutoff_7_5_df{year}']
    xr = pd.to_numeric(res_df['PP RANK'], errors='coerce')
    yr = pd.to_numeric(res_df['OC'], errors='coerce')
    mask = xr.notna() & yr.notna()
    xr = xr[mask]
    yr = yr[mask]
    coef_res = np.polyfit(xr, yr, 1)
    poly1d_res = np.poly1d(coef_res)
    plt.plot(xr, poly1d_res(xr), linestyle='--', color=res_colors[i],
             label=f'7.5% Reservation Trend {year}')

    # Intersection
    m1, c1 = coef_gen
    m2, c2 = coef_res
    if m1 != m2:
        x_int = (c2 - c1) / (m1 - m2)
        y_int = m1 * x_int + c1
        plt.scatter(x_int, y_int, color=intersect_colors[i], zorder=5)
        plt.plot(x_int, y_int, 'o', color=intersect_colors[i])
        plt.legend().set_title("Legend")
        plt.plot([], [], 'o', color=intersect_colors[i],
                 label=f'Intersection {year}: (Rank-{x_int:.1f}, Cutoff-{y_int:.1f})')

# Final plot details
plt.xlabel('Public Perception Rank')
plt.ylabel('Cutoff')
plt.title('OC Cutoff: General vs 7.5% Reservation (2022–2024)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
